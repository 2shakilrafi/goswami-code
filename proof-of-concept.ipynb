```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import mpl_toolkits
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=3000,
                  n_features= 9,
                  centers=8,
                  cluster_std=1.0,random_state=101)
```

```python
data
```

```python
X,y = data
```

```python
X
```

```python
y
```

```python
np.random.seed(seed = 101)
z_noise = np.random.normal(size = len(X))
z_noise = pd.Series(z_noise)
```

```python
z_noise
```

```python
feat = pd.DataFrame(X)
```

```python
feat
```

```python
feat = pd.concat([feat,z_noise],axis=1)
```

```python
feat.columns = ['X1','X2','X3','X4','X5','X6','X7','X8','X9','X10']
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from tensorflow.keras.optimizers import SGD
```

```python
# 10 --> 7 --> 3 --> 5 --> 10
```

```python
encoder = Sequential()
encoder.add(Dense(units=7, activation='relu',input_shape =[10]))
encoder.add(Dense(units=3, activation='relu',input_shape = [7]))
```

```python
decoder = Sequential()
decoder.add(Dense(units=5, activation='relu', input_shape = [3]))
decoder.add(Dense(units=10, activation='relu', input_shape = [5]))
```

```python
autoencoder = Sequential([encoder, decoder])
```

```python
autoencoder.compile(loss = 'mse',
                    optimizer = SGD(learning_rate= 1.0))
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(feat)
```

```python
scaled_data
```

```python
autoencoder.fit(scaled_data, 
                scaled_data, 
                batch_size = 256,
                shuffle = True,
                epochs = 10)
```

```python
encoded_2dim = encoder.predict(scaled_data)
```

```python
encoded_2dim
```
